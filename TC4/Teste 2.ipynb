```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Carregar os dados
url = "sensor_readings_24.data"
data = pd.read_csv(url, header=None)

# Pré-processamento: Normalizar os dados
scaler = StandardScaler()
normalized_data = scaler.fit_transform(data.iloc[:, :-1])

# Aplicar o algoritmo de k-médias (vamos usar k=3 como exemplo)
kmeans = KMeans(n_clusters=10, random_state=42)
kmeans.fit(normalized_data)

# Adicionar os rótulos dos clusters ao DataFrame
data['Cluster'] = kmeans.labels_

# Exibir os resultados da clusterização
print("Resultados da Clusterização:")
print(data[['Cluster', 24]].value_counts())

# Visualizar os clusters (usando as duas primeiras colunas para simplificar)
plt.scatter(normalized_data[:, 0], normalized_data[:, 1], c=data['Cluster'], cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, marker='X', c='red', label='Centroids')
plt.title("Clusters Identificados pelo K-Means")
plt.xlabel("Atributo 0 (Normalizado)")
plt.ylabel("Atributo 1 (Normalizado)")
plt.legend()
plt.show()
```
